```python
combined = vol_reg
combined['availability'] = combined[times].apply(lambda x: ''.join(x).strip(), axis=1)
combined = combined.drop(times, axis=1)
combined = combined[['email', 'availability', 'hours', 'first_pref', 'second_pref', 'third_pref']]
# combined
```

```python
mapping = combined.set_index('email').to_dict()
availability = mapping['availability']
hours = mapping['hours']
first_pref = mapping['first_pref']
second_pref = mapping['second_pref']
third_pref = mapping['third_pref']
pref = {k:[v, second_pref[k], third_pref[k]] for k,v in first_pref.items()}
```

```python
def compute_hrs(slot):
    time = slot.split('_')[2]
    split = time.split('-')
    start = int(split[0])
    end = int(split[1])
    return end - start

def generate_hrs(k, v):
    avail = availability[k].split('|')
    avail = filter(None, avail)
    ret = [(a, compute_hrs(a)) for a in avail]
    return (v, pref[k], ret)
    

volunteers = {k: generate_hrs(k,v) for k, v in hours.items()}
pp.pprint(volunteers)
```

```python
stations = pd.read_csv('stations.csv')
stations = stations.set_index('Stations')
stations.columns = times_filtered
stations = stations.T
stations_dict = stations.to_dict()

def remove_empty_slots(slot_dict):
    return {k: v for k, v in slot_dict.items() if v != 0 }

stations = {k: remove_empty_slots(v) for k,v in stations_dict.items()}
# pp.pprint(stations)
```

```python
for station, time_slots in stations.items():
    for time_slot, capacity in time_slots.items():
        stations[station][time_slot] = {'capacity':capacity, 'volunteers':set()}
pp.pprint(stations)
```

```python
def add_vol_to_slot(vol, station, time_slot):
    if stations[station][time_slot]['capacity'] >= len(volunteers)
        return False
    
    stations[station][time_slot]['volunteers'].add(vol)
    
def remove_vol_from_slot(vol, station, time_slot):
    if stations[station][time_slot]['capacity'] <= 0 or vol not in stations[station][time_slot]['volunteers']
        return False
    
    stations[station][time_slot]['volunteers'].remove(vol)
```
